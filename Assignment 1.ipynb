# Derivatives

**Definition of a Derivative:**

$ \frac{d}{{dx}}f\left( x \right) = \mathop {\lim }\limits_{\Delta \to 0} \frac{{f\left( {x + \Delta } \right) - f\left( x \right)}}{\Delta } $





## Geometric Interpretation of a Derivative



As illustrated, the specific slope at a single point would be impossible to find without a derivative. The geometric definition allows a person or application to know the exact value at a given point. This is also known as the instantaneous rate of change. Estimations are too slow and inaccurate when compared to taking the derivative.

## Practical Uses of Derivatives

- instantaneous velocity
- machine learning
- slope of a line at a specific point
- rate of change

Derivatives allow for several advances in science. For example, many laws about motion could not have been discovered if not for the derivative. Instantaneous velocity would be impossible to determine without the derivative. In machine learning, many advances have only been possible with the help of the derivative. Optimization in Machine Learning would be impossible without the help of algorithms which utilize derivatives.
